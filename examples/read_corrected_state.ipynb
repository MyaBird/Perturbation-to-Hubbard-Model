```python
import perturbation_to_hubbard_model
from perturbation_to_hubbard_model import InitializeCircuit
from perturbation_to_hubbard_model import HubbardCircuit
import pennylane as qml
import numpy as np

# Initialize circuit
cir_init = InitializeCircuit(4,2)
dev = cir_init.create_device('default.qubit')
shots = 4

# Compile circuit gates
@qml.set_shots(shots=shots)
@qml.qnode(dev)
def circuit(config, N, M, pert_coeff, transfer, interaction, theta, phase, c, e_gs, e_neg1, e_0, e_1, e_2, e_h):
    HubbardCircuit(config, N, M, pert_coeff, transfer, interaction, theta, phase, c, e_gs, e_neg1, e_0, e_1, e_2, e_h).build_circuit()

    qubits = []
    for i in range(N):
        qml.PauliZ(wires=cir_init.qubits[i])
        qubits.append(cir_init.qubits[i])
    
    qml.PauliZ(wires="q1''")
    qubits.append("q1''")

    sample_output = []
    for i in qubits:
        sample_output.append(qml.sample(wires=i))

    return sample_output

# Define variables:
config = [0,0,0,1]
N = 4
M = 2
pert_coeff = 0.001
transfer = 1
interaction = 1
theta = np.pi/4
phase = 1
c = 0.5
e_gs = -1.56
e_neg1 = -1
e_0 = 0
e_1 = 1
e_2 = 2
e_h = 2.56

# Print output of circuit:
output = circuit(config, N, M, pert_coeff, transfer, interaction, theta, phase, c, e_gs, e_neg1, e_0, e_1, e_2, e_h)

for j in range(shots):
    readout = []
    for q in range(N):
        readout.append(int(output[q][j][0]))
    print(f"Trial {j+1}: {"-" * 20}")
    print(f"Output config: {readout}")
    print(f"Readout qubit: {int(output[N][j][0])}")
    print("\n")
```
